```python
# actual code for project data

url = "https://api.yelp.com/v3/businesses/search"
api_key = "w38fgOOsrJOAuyQvVeZl9Po7cvQT2JicNX-81_I7nlb3muDjKEUgYyExyhFPn0cIaopDSBPuQriaIRhL6nteZqPtaJNqPJCfpHCzFM8jKBVpAW9Um2jsE13UA_LuXXYx"
headers = {'Authorization': 'Bearer {}'.format(api_key),}
term = 'Fast Food'
locations = ['Raleigh, US','Charlotte, US','Durham, US']
SEARCH_LIMIT = 50
food_rawdata = []
counter = 0
total = 0

for location in locations:
        url_params = {
                'term': term.replace(' ', '+'),
                'location': location.replace(' ', '+'),
                'limit': SEARCH_LIMIT
                    }
        response = requests.get(url, headers=headers, params=url_params).json()
        for responses in response ["businesses"]:
             food_rawdata.append(responses)
            
for food_data in food_rawdata:
    total = total + 1
    
print(total)
    
```
